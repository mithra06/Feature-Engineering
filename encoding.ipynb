# mean encoding

```python
df1=pd.read_csv('cars.csv')
```

```python
df1.head()
```

```python
mean=df1.groupby([' brand'])[' year'].mean().to_dict()
```

```python
mean
```

```python
df1[' brand']=df1[' brand'].map(mean)
```

```python
df1.head()
```

# target guided

```python
df2=pd.read_csv('cars.csv')
```

```python
df2.head()
```

```python
tar=df2.groupby([' brand'])[' year'].mean().sort_values().index
```

```python
tar
```

```python
enumerate(tar,0)
```

```python
ord={k:i for i,k in enumerate(tar,0)}
    
```

```python
ord
```

```python
df2[' brand']=df2[' brand'].map(ord)
```

```python
df2.head()
```

# one hot encoder

```python
df4=pd.read_csv('cars.csv')
```

```python
df4.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse=False)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
values=df4[' brand'].values
```

```python
ct=ColumnTransformer([('encoder',ohe,[0])],remainder='passthrough')
df4=np.array(ct.fit_transform(df4),dtype=np.str)
```

```python
df4=pd.DataFrame(df4)
```

```python
df4.head()
```

```python
df4=pd.get_dummies(df4,drop_first=True)
```
